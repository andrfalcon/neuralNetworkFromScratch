```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
# Neural network for a single observation
x = np.array(df.iloc[0, 1:])
y = df.iloc[0, 0]
```

```python
class NeuralNetwork:
    # num_l: 0, 1, 2, 3
    # nodes_l: [784, 3, 2, 1]
    def __init__(self, X, y, num_l, nodes_l,):
        self.X = X
        self.y = y
        self.num_l = num_l
        self.nodes_l = nodes_l
        self.weights = {i: np.array([list(np.ones(self.nodes_l[i-1])) for j in range(self.nodes_l[i])]) for i in range(1, num_l+1)}
        self.biases = {i:np.ones(self.nodes_l[i]) for i in range(1, self.num_l + 1)}

    
    # Returns the output of the neural network after a single feed-forward iteration
    def feed_forward(self):
        a = relu((self.weights[1] @ self.X) + self.biases[1])
        for i in range(2, self.num_l + 1):
            a = (self.weights[i] @ a) + self.biases[i]
        return a[0]
        
    def relu(v):
        return np.array([max(0, i) for i in v])
            
            
        
```

```python
my_nn = NeuralNetwork(x, y, 3, [784, 3, 2, 1])
```
